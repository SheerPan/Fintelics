## Suppose you already have a Fintelics_Image folder to store all your images in the first part. Now you should create another folder named "Fintelics_features" under the root folder of YOUR OWN Google drive.

```python
folder_name = "Fintelics_Image" # Your image folder name
csv_folder_name = "Fintelics_features" # The place you store csv files 
csv_file_name = "40000_60000.csv" # You can change the name
csv_path = "/content/drive/MyDrive/"+ folder_name +"/"+ csv_folder_name + "/" + csv_file_name
```

```python
# Assume that the image is in RGB format
input_layer = tf.keras.layers.Input(shape=(32, 32, 3))

# Padding == 1
conv_layer = tf.keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', activation='relu')(input_layer)

# Size = 2 * 2
pool_layer = tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv_layer)
flat_layer = tf.keras.layers.Flatten()(pool_layer)

# Add a dense layer with 512 units and ReLU activation
dense_layer = tf.keras.layers.Dense(512, activation='relu')(flat_layer)

model = tf.keras.models.Model(inputs=input_layer, outputs=dense_layer)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
array = glob.glob("/content/drive/MyDrive/" + folder_name +"/Fintelic_Image"+ "/*.png") # get all images (.png) from your drive 
print(len(array))
print(array[0])
```

```python
def create_csv():
  with open(csv_path, 'w', newline='') as outcsv:
    writer = csv.writer(outcsv)
    cols = ["feature" + str(i) for i in range(1, 513, 1)]
    cols.insert(0, "index")
    writer.writerow(cols)
  return cols
```

```python
def write_csv(index:str, features, field_names):
  array = [str(features[0, i]) for i in range(0, 512)]
  array.insert(0, index)
  row_dict = {}
  for i in range(512):
    row_dict[field_names[i]] = array[i]
  with open(csv_path, 'a') as csv_file:
    dict_object = csv.DictWriter(csv_file, fieldnames=field_names) 
    dict_object.writerow(row_dict)
```

```python
# Input your csv file folder name 
def convert_image_to_features(field_names):
  cnt = 0
  for i in range(len(array)):
    path = array[i]
    index = path.replace("/content/drive/MyDrive/" + folder_name + "/Fintelic_Image" + "/", "").replace(".png", "")
    try:
      image = Image.open(path).convert("RGB")
      image = np.reshape(image, (1, 32, 32, 3))
      features = model.predict(image)
      write_csv(index, features, field_names)
      print(index + ": Work done successfully")
      cnt += 1
    except Exception as e:
      print(e)
      continue
  return cnt

```

```python
field_names = create_csv()
cnt = convert_image_to_features(field_names)
print("Number of images processed: " + str(cnt))
print("Successful rate: " + str(cnt / len(array)))
```
